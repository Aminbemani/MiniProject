```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df_train=pd.read_csv('mnist_train.csv')
df_test=pd.read_csv('mnist_test.csv')
df_train=df_train.to_numpy()
df_test=df_test.to_numpy()
X=df_train[:,1:]
y=df_train[:,0]
X=X/255
X_train, X_val, y_train, y_val=train_test_split(X,y, test_size=0.2)
```

```python
train_scores = []
Validation_scores = []
for c in np.arange(0.1, 10, 0.1):
    clf = LogisticRegression(C=c, penalty='l2')
    clf.fit(X_train, y_train)
    
    train_scores.append(clf.score(X_train, y_train))
    Validation_scores.append(clf.score(X_val, y_val))

plt.plot(np.arange(0.1, 10, 0.1), np.array(train_scores), 'b-')
plt.plot(np.arange(0.1, 10, 0.1), np.array(Validation_scores), 'r-')
```

```python
clf = LogisticRegression(C=0.3, penalty='l2')
clf.fit(X_train, y_train)
X_test=df_test[:,1:]
y_test=df_test[:,0]
X_test=X_test/255  
print('test score:', clf.score(X_test, y_test))
```
